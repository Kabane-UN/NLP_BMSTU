```python
tags = ["ADJ", "N", "V", "CONJ", "DET"]

transition = {
    "START": {"ADJ": 0.2, "N": 0.2, "V": 0.1, "CONJ": 0.01,"DET": 0.4},
    "ADJ": {"ADJ": 0.2, "N": 0.5, "V": 0.05, "CONJ": 0.2,"DET": 0.01},
    "N": {"ADJ": 0.05, "N": 0.01, "V": 0.5, "CONJ": 0.2,"DET": 0.01},
    "V": {"ADJ": 0.1, "N": 0.2, "V": 0.01, "CONJ": 0.2,"DET": 0.3},
    "CONJ": {"ADJ": 0.1, "N": 0.2, "V": 0.2, "CONJ": 0.0,"DET": 0.2},
    "DET": {"ADJ": 0.3, "N": 0.7, "V": 0.0, "CONJ": 0.0,"DET": 0.0},
}

emission = {
    "ADJ": {"arrow": 0.01, "time": 0.0005},
    "N": {"time": 0.01, "arrow": 0.01, "flies": 0.0005, "like": 0.001},
    "V": {"flies": 0.01, "like": 0.02, "time": 0.001},
    "CONJ": {"like": 0.05},
    "DET": {"an": 0.1},
}

sentence = ["time", "flies", "like", "an", "arrow"]
```

```python
V = [{}]
backpointer = [{}]

for t in tags:
    V[0][t] = transition["START"].get(t, 0) * emission[t].get(sentence[0], 0)
    backpointer[0][t] = None
V, backpointer
```

```python
for i in range(1, len(sentence)):
    V.append({})
    backpointer.append({})
    for t in tags:
        (prob, prev_state) = max(
            (V[i-1][t_prev] * transition[t_prev].get(t, 0) * emission[t].get(sentence[i], 0), t_prev)
            for t_prev in tags
        )
        V[i][t] = prob
        backpointer[i][t] = prev_state

last_tag = max(V[-1], key=V[-1].get)
V, backpointer
```

```python
best_path = [last_tag]
for i in range(len(sentence)-1, 0, -1):
    best_path.insert(0, backpointer[i][best_path[0]])

print("Предложение:", " ".join(sentence))
print("Теги:", best_path)
```
